```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Student_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

drop Unnamed column

```python

df = df.drop(columns=["Unnamed:0"])
df
```

Change weekly study hours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df
```

 Gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analysed that:
the number of females in the data is more than the number of males

parent Education impact on student score

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean','ReadingScore':'mean','WritingScore':'mean'})
gb
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship Between parent's education and student score")
plt.show()
```

from the above chart we have conclude that the education of the parent have
a good impact on student score

Parent Marital status impact on student scores

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean','ReadingScore':'mean','WritingScore':'mean'})
gb1
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parents Marital status and students scores")
plt.show()
```

from the above chart we have analyze that marital status of parent have 
neglible impact on student score

Distribution of Ethnic Groups

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.figure(figsize=(6,6))
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.legend()
plt.show()
print(mlist)
```
